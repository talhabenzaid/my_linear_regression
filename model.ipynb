```python
import numpy as np
import matplotlib.pyplot as plt
from Linear_regression import LinearRegression 


X = 2 * np.random.rand(100, 1)
Y = 3 * X.flatten() + 4 + np.random.randn(100)


plt.scatter(X, Y, color='blue', label='Data points')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
model = LinearRegression(lr=0.1, n_iters=1000)
model.fit(X, Y)


X_line = np.linspace(0, 2, 100).reshape(100, 1)
Y_line = model.predict(X_line)


plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X_line, Y_line, color='red', label='Regression line')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python

X_new = np.array([[1.5], [0.5], [2.0]])
Y_new_pred = model.predict(X_new)

print("Predictions for new inputs:")
for x_val, y_pred in zip(X_new.flatten(), Y_new_pred):
    print(f"X={x_val} => Predicted Y={y_pred:.2f}")

plt.scatter(X, Y, color='blue', label='Training data')
plt.scatter(X_new, Y_new_pred, color='green', label='Predictions')
plt.plot(np.linspace(0, 2, 100), model.predict(np.linspace(0, 2, 100).reshape(100,1)), color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression with Random Data')
plt.show()
```
